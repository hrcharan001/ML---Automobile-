```python
# we are separating the feature column with x and target column with y
x = df
y = df['price']
```

```python
# now from the features column we are separating it with numerical and categorical types
feature = x.drop('price',axis = 1)
num=feature.select_dtypes(include=['int64', 'float64']).columns
cat=x.select_dtypes(exclude=['int64','float64']).columns

# we are doing standard scalar for numerical columns as it can handle outliers well comparing to min-max scaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x[num] = scaler.fit_transform(x[num])

# since sum columns can be renamed the word to numbers we are using the map function to assign the values to the numbers and converting it into numerical values
# this is called as manual mapping
x['num-of-doors'] = x['num-of-doors'].map({
    'two': 2,
    'four': 4
})

x['num-of-cylinders'] = x['num-of-cylinders'].str.strip()

x['num-of-cylinders'] = x['num-of-cylinders'].map({
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'eight': 8,
    'twelve': 12
})
```

```python
x
```

```python
# we are converting categorical data to numerical data by using label encoder
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

x['fuel-type'] = le.fit_transform(x['fuel-type'])
```

```python
x['engine-location'] = le.fit_transform(x['engine-location'])
```

```python
x['aspiration'] = le.fit_transform(x['aspiration'])
```

```python
corr = x.select_dtypes(include='number').corr()

corr
# sns.heatmap(corr,annot=True)
```

```python
corr_matrix = x.corr(numeric_only=True)

corr_pairs = (
    corr_matrix
    .unstack()
    .sort_values(key=abs, ascending=False)
)

# Keep only correlations > 0.75 or < -0.75
high_corr = corr_pairs[(abs(corr_pairs) > 0.75) & (abs(corr_pairs)<1)]

print(high_corr)
```

```python
# feature selection
# we are removing some columns beacause of high correlation among themselves, so we are dropping the columns to avoid redundancy of data in the columns 
x = x.drop(columns=['compression-ratio','city-mpg','length','width','curb-weight','horsepower'])
```

```python
x = x.drop(columns=['price'])
```

```python
x.columns
```

```python
# encoding
# # we are converting categorical values to numerical values using one hot encoding
one_hot_cols = [
     'make',
     'body-style',
     'drive-wheels',
     'engine-type',
     'fuel-system'
]

x = pd.get_dummies(
    x,
    columns=one_hot_cols,
    drop_first=True
)
```

```python
x
```

```python
pd.reset_option("display.max_rows",None)
```

```python
x.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Model selection
le = LinearRegression()

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
le.fit(x_train,y_train)
y_pred=le.predict(x_test)
```

```python
# checking for errors with error metrics(regression metrics) and R2 score
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error,r2_score
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(root_mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))


```

```python
# to check the input and output(predicted) value by taking 1DP from input and checking it with corresponding output
sample = x_test.iloc[[11]]
actual = y_test.iloc[11]

pred = le.predict(sample)

print("Actual value:", actual)
print("Predicted value",pred[0])
print("Input:\n", sample)
```

```python
# lets predict usinng Decision Tree
from sklearn.tree import DecisionTreeRegressor

dc = DecisionTreeRegressor(max_depth=5,min_samples_split=5,random_state=42)

dc.fit(x_train,y_train)
dc_pred=dc.predict(x_test)
```

```python
 # finding the error metrics for decision treecd
print(mean_squared_error(y_test,dc_pred))
print(mean_absolute_error(y_test,dc_pred))
print(root_mean_squared_error(y_test,dc_pred))
print(r2_score(y_test,dc_pred))
dc_train_pred = dc.predict(x_train)
print("Train R2 score:", r2_score(y_train,dc_train_pred))
dc_pred
```
